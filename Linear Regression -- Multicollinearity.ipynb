## Scaling the dataset

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
Y = data["Sale_Price"]
X = scaler.fit_transform(data.drop(columns=["Sale_Price"]))
X = pd.DataFrame(data = X, columns = data.drop(columns=["Sale_Price"]).columns)
X.head()
```

## Calculating VIF

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor


vif_data = X

VIF = pd.Series([variance_inflation_factor(vif_data.values, i) for i in range(vif_data.shape[1])], index = vif_data.columns)
VIF
```

```python
def MC_remover(data):
    vif = pd.Series([variance_inflation_factor(data.values,i) for i in range(data.shape[1])], index = data.columns)
    
    if vif.max()>5:
        print(vif[vif==vif.max()].index[0],"has been removed")
        data = data.drop(columns = [vif[vif == vif.max()].index[0]])
        return data
    else:
        print("NO multicollinearity present anymore")
        return data
```

```python
for i in range(7):
    vif_data = MC_remover(vif_data)
    
vif_data.head()
```

## Remaining Columns 

```python
## Calculating VIF for remainig columns

VIF = pd.Series([variance_inflation_factor(vif_data.values,i) for i in range(vif_data.shape[1])], index = data.columns)
VIF,len(vif_data.columns)
```

## Train Test set

```python
x = vif_data
y = data["Sale_Price"]
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(X,Y, test_size = 0.3, random_state = 101)

xtrain.shape,xtest.shape,ytrain.shape,ytest.shape
```

## Linear Regression

```python
from sklearn.linear_model import LinearRegression

lr = LinearRegression(normalize = True)

lr.fit(xtrain, ytrain)
```

```python
lr.coef_
```

```python
predictions = lr.predict(xtest)
lr.score(xtest,ytest)
```

```python
residuals = predictions - ytest
residuals_table = pd.DataFrame({"residuals":residuals,"predictions":predictions})
residuals_table = residuals_table.sort_values(by = "predictions")
```

```python
z = [i for i in range(int(residuals_table["predictions"].max()))]
k = [0 for i in range(int(residuals_table["predictions"].max()))]
```

```python
plt.figure(dpi=130, figsize=(17,7))

plt.scatter(residuals_table["predictions"], residuals_table["residuals"], color = "red", s = 2 )
plt.plot(z,k, color = "green", linewidth=3,label="regression line")
plt.ylim(-800000,800000)
plt.xlabel("fitted points")
plt.ylabel("residual plot")
plt.legend()
plt.show()
```

## Distribution of errors

```python
plt.figure(dpi=100, figsize=(10,7))

plt.hist(residuals_table['residuals'],color = "red", bins=200)
plt.xlabel("residuals")
plt.ylabel("frequency")
plt.title("distribution of residuals")
plt.show()
```

## Model coefficients

```python
coefficients_table = pd.DataFrame({"column":xtrain.columns, "coefficients":lr.coef_})
coefficient_table = coefficients_table.sort_values(by="coefficients")
```

```python
plt.figure(figsize=(8,6), dpi=120)
x = coefficient_table["column"]
y = coefficient_table["coefficients"]
plt.barh(x,y)
plt.xlabel("coefficints")
plt.ylabel("variables")
plt.title("Normalised coefficient plot")
plt.show()
```
